# Популярные имена

```python
def extract_name(full_name):
    parts = full_name.split(', ')[1].split('. ')
    return parts[1].strip('"') if len(parts) > 1 else full_name

df['FirstName'] = df['Name'].apply(extract_name)

male_names = df[df['Sex'] == 'male']['FirstName']
female_names = df[df['Sex'] == 'female']['FirstName']

print(f"Самое популярное мужское имя: {male_names.mode()[0]}")
print(f"Самое популярное женское имя: {female_names.mode()[0]}")
```

# Популярные имена по классам

```python
for pclass in sorted(df['Pclass'].unique()):
    print(f"\nКласс {pclass}:")
    class_df = df[df['Pclass'] == pclass]

    male_in_class = class_df[class_df['Sex'] == 'male']['FirstName']
    female_in_class = class_df[class_df['Sex'] == 'female']['FirstName']

    print(f"  Мужское: {male_in_class.mode()[0]}")
    print(f"  Женское: {female_in_class.mode()[0]}")
```

# Кол-во n-местных кабин

```python
cabin_data = df['Cabin'].dropna()
all_cabins = []

for cabins in cabin_data:
    if isinstance(cabins, str):
        all_cabins.extend(cabins.split())

cabin_counts = pd.Series(all_cabins).value_counts()
n_berth_counts = cabin_counts.value_counts().sort_index()

print(n_berth_counts[n_berth_counts.index >= 2])
```
